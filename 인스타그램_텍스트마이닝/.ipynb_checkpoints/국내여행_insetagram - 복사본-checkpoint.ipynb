```python
import re
tag2 = []
for i in range(0,len(tag)):
    tag1 = re.findall(r'#[0-9,가-힣,a-z]+',tag[i])
    tag2.append(tag1)
```

```python
tag2 = sum(tag2,[])
```

```python
tag_re = []
for i in range(0,len(tag2)):
    a = re.sub(r'#','',tag2[i])
    tag_re.append(a)
```

```python
def write_txt(list, fname, sep):
    file = open(fname,'w')
    vstr = ''
    
    for a in list:
        vstr = vstr + str(a) + sep
    vstr = vstr.rstrip(sep)
    
    file.writelines(vstr)
    file.close()
    print("[파일 저장 완료]")
```

```python
write_txt(tag_re, 'korea_travel.txt',sep=',')
```

```python
file = open('korea_travel.txt','r')
tag_all = file.read()
tag_all
```

```python
tag = tag_all.split(',')
```

```python
# 단어의 빈도수 확인
from collections import Counter
countss = Counter(tag)
countss
# 딕셔너리로 변환
word_dict = dict(countss)
word_dict
```

```python
from wordcloud import WordCloud
wc = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf', background_color='white',\
               width=400, height=200)\
                .generate_from_frequencies(word_dict)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wc)
plt.show()
```
